```python
import codecademylib
import pandas as pd

visits = pd.read_csv('visits.csv',
                     parse_dates=[1])
cart = pd.read_csv('cart.csv',
                   parse_dates=[1])
checkout = pd.read_csv('checkout.csv',
                       parse_dates=[1])
purchase = pd.read_csv('purchase.csv',
                       parse_dates=[1])

print(visits.head())
print(cart.head())
print(checkout.head())
print(purchase.head())

visits_cart = pd.merge(visits, cart, how="left")
print(visits_cart.head(10))

visits_cart_len = len(visits_cart)
print(visits_cart_len)

cart_time_null = len(visits_cart[visits_cart.cart_time.isnull()])
print(cart_time_null)

not_placing_in_cart = cart_time_null/float(visits_cart_len)
print(not_placing_in_cart)


cart_checkout = pd.merge(cart, checkout, how="left")
print(cart_checkout.head(10))

cart_checkout_len = len(cart_checkout)
print(cart_checkout_len)

cart_checkout_null = len(cart_checkout[cart_checkout.checkout_time.isnull()])
print(cart_checkout_null)

not_placing_in_cart2 = cart_checkout_null/float(cart_checkout_len)
print(not_placing_in_cart2)


all_data = visits.merge(cart).merge(checkout).merge(purchase)
print(all_data.head())

checkout_purchase = pd.merge(checkout, purchase, how="left")
checkout_purchase_len = len(checkout_purchase)
print(checkout_purchase_len)
checkout_purchase_null = len(checkout_purchase[checkout_purchase.purchase_time.isnull()])
print(checkout_purchase_null)
not_placing_in_cart3 = checkout_purchase_null/float(checkout_purchase_len)
print(not_placing_in_cart3)


weakest = "visits_cart"


all_data['time_to_purchase'] = all_data.purchase_time - all_data.visit_time


print(all_data.time_to_purchase)
print(all_data.time_to_purchase.mean())
```
